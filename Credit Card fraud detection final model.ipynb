## There is one more way to determine the presence of outlier in the given data

```python
# Create a function to return index of outliers 
def indicies_of_outliers(x): 
    q1, q3 = np.percentile(x, [25, 75]) 
    iqr = q3 - q1 
    lower_bound = q1 - (iqr * 3) 
    upper_bound = q3 + (iqr * 3)
    return np.where((x > upper_bound) | (x < lower_bound))
```

# Under Sampling

```python
from collections import Counter # counter takes values returns value_counts dictionary
from sklearn.datasets import make_classification

print('Original dataset shape %s' % Counter(y))

rus = RandomUnderSampler(random_state=1)

X_res, y_res = rus.fit_resample(X, y)
print('Resampled dataset shape %s' % Counter(y_res))
```

### AS we can see above the data is under sampled now the problem of class imbalance is resolved

### Now we will do all the steps done above again

```python
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.3, random_state=1)
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
y_pred = logreg.predict(X_test)

from sklearn.metrics import accuracy_score
print(accuracy_score(y_pred , y_test))  # Accuracy is surely reduced , let's look at the roc curve now
```

```python
y_pred_proba = logreg.predict_proba(X_test)[::,1]
fpr, tpr, _ = roc_curve(y_test,  y_pred)
auc = roc_auc_score(y_test, y_pred)
plt.plot(fpr,tpr,label="data 1, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

### As we can see now the AUC ROC score and accurecy score using logistic regression has now improved
